```python
from pyspark.sql import functions as F

data = data.filter((F.col('trip_distance') > 0) &
                   (F.col('fare_amount') > 0) &
                   (F.col('total_amount') > 0) &
                   (F.col('tpep_pickup_datetime').isNotNull()) &
                   (F.col('tpep_dropoff_datetime').isNotNull()))
data = data.sample(fraction=0.80, seed=42)
```

```python
data.count()
```

```python
count_nulls = data.select([F.sum(F.col(c).isNull().cast('int')).alias(c) for c in data.columns])
count_nulls.show()
```

```python
data = data.fillna({'passenger_count': -1})
```

```python
from pyspark.sql.functions import col, to_timestamp

data = data.withColumn('tpep_pickup_datetime', to_timestamp('tpep_pickup_datetime'))
data = data.withColumn('tpep_dropoff_datetime', to_timestamp('tpep_dropoff_datetime'))
data.show(2)
```

```python
from pyspark.sql.functions import date_trunc, count, sum

hourly_data = data.withColumn('pickup_hour', date_trunc('hour', col('tpep_pickup_datetime'))) \
    .groupBy('pickup_hour') \
    .agg(
        count('*').alias('trip_count'),
        sum('total_amount').alias('total_revenue'),
        sum('trip_distance').alias('total_distance')
    )
```

```python
daily = hours.resample('D').sum()
daily = daily.sort_index(ascending=False)

daily['rolling_mean'] = daily['trip_count'].rolling(30).mean()

plt.figure(figsize=(16,7))
plt.plot(daily.index, daily['trip_count'], label='Original', alpha=0.5)
plt.plot(daily.index, daily['rolling_mean'], label='Mean', color='red')
plt.legend()
plt.show()
```

```python
def feature(df):
    df['hour'] = df.index.hour
    df['day'] = df.index.dayofyear
    df['dayofweek'] = df.index.dayofweek
    df['month'] = df.index.month
    df['quarter'] = df.index.quarter
    df['year'] = df.index.year

    return df
```

```python
from sklearn.metrics import mean_squared_error
from itertools import product

param_grid = {
    'max_depth': [9, 10, 11],
    'min_child_weight': [3, 4],
    'subsample': [0.8, 0.9],
    'colsample_bytree': [0.9, 1],
    'reg_alpha': [0.6, 0.7, 0.8],
    'reg_lambda': [6, 7, 8]
}

all_combinations = list(product(
    param_grid['max_depth'],
    param_grid['min_child_weight'],
    param_grid['subsample'],
    param_grid['colsample_bytree'],
    param_grid['reg_alpha'],
    param_grid['reg_lambda']
))

best_rmse = float("inf")
best_params = None

for params in all_combinations:
    max_depth, min_child_weight, subsample, colsample_bytree, reg_alpha, reg_lambda = params

    model = xgb.XGBRegressor(
        n_estimators=3000,
        learning_rate=0.05,
        max_depth=max_depth,
        min_child_weight=min_child_weight,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        reg_alpha=reg_alpha,
        reg_lambda=reg_lambda,
        early_stopping_rounds=50
    )

    model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)

    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    if rmse < best_rmse:
        best_rmse = rmse
        best_params = params

    print(f"Max_depth={max_depth}, Min_child_weight={min_child_weight}, Subsample={subsample}, Colsample_bytree={colsample_bytree}, Reg_alpha={reg_alpha}, Reg_lambda={reg_lambda} RMSE: {rmse:.4f}")

print("\nBest RMSE:", best_rmse)
print("Best Hyperparameters:", {
    'max_depth': best_params[0],
    'min_child_weight': best_params[1],
    'subsample': best_params[2],
    'colsample_bytree': best_params[3],
    'reg_alpha': best_params[4],
    'reg_lambda': best_params[5]
})
```

```python
from sklearn.preprocessing import MinMaxScaler
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
hours = hours['trip_count'].to_frame()
hours.head()
```

```python
def create_sequences(data, steps):
    for i in range(1, steps+1):
        data[f'Close(t-{i})'] = data['trip_count'].shift(i)

    data.dropna(inplace=True)
    return data

lookback = 24
```

```python
hours = hours.sort_index()
train_data = hours.loc['2011-01-01':'2023-12-31']
test_data = hours.loc['2024-01-01':'2024-11-30']
```

```python
train_data.head()
```

```python
train_transformed = create_sequences(train_data.copy(), lookback)
test_transformed = create_sequences(test_data.copy(), lookback)
```

```python
train_transformed.head()
```

```python
X = np.concatenate([train_transformed.drop(columns=['trip_count']).values,
                    test_transformed.drop(columns=['trip_count']).values])

y = np.concatenate([train_transformed['trip_count'].values,
                    test_transformed['trip_count'].values])
```

```python
X
```

```python
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()
```

```python
split = len(train_transformed)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
X_train = X_train.reshape((-1, lookback, 1))
X_test = X_test.reshape((-1, lookback, 1))
y_train = y_train.reshape((-1, 1))
y_test = y_test.reshape((-1, 1))
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float32).to(device)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).to(device)

X_test_tensor = torch.tensor(X_test, dtype=torch.float32).to(device)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).to(device)
```

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
class TimeDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, item):
        return self.X[item], self.y[item]
```

```python
train_dataset = TimeDataset(X_train_tensor, y_train_tensor)
test_dataset = TimeDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(LSTM, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        batch_size = x.size(0)

        h0 = torch.zeros(self.num_layers, batch_size, self.hidden_size).to(device)
        c0 = torch.zeros(self.num_layers, batch_size, self.hidden_size).to(device)

        out, _ = self.lstm(x, (h0, c0))

        out = self.fc(out[:, -1, :])
        return out
```

```python
model = LSTM(1, 64, 1).to(device)
```

```python
loss_function = nn.MSELoss()

learning_rate = 0.001
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
def train_model(model, train_loader):
    model.train()

    for epoch in range(20):
        running_loss = 0.0

        for batch in train_loader:
            x_batch, y_batch = batch[0].to(device), batch[1].to(device)

            output = model(x_batch)
            loss = loss_function(output, y_batch)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        avg_loss = running_loss / len(train_loader)
        print(f'Epoch [{epoch + 1}/20], Loss: {avg_loss:.4f}')
```

```python
train_model(model, train_loader)
print('Done')
```

```python
model.eval()

with torch.no_grad():
    y_pred = model(X_test_tensor.to(device))
    y_pred = y_pred.cpu().numpy()

y_test = y_test_tensor.cpu().numpy()
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model(X_test_tensor).cpu().detach().numpy().flatten()
y_true = y_test_tensor.cpu().numpy()

mse = mean_squared_error(y_true, y_pred)
rmse = mse ** 0.5
print(f"Test RMSE: {rmse:.4f}")
```
